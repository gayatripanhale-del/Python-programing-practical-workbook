<h2>Question Set A

<h3>1) Write a program to find the union, intersection, and difference of two sets entered by the
user.</h3>

```python
s1 = set(input("Enter elements of first set (no spaces): "))
s2 = set(input("Enter elements of second set (no spaces): "))

print("Union:", s1 | s2)
print("Intersection:", s1 & s2)
print("Difference (s1 - s2):", s1 - s2)
```

<h3>2) Write a program to check if a set is a subset of another set.

```python
s1 = set(input("Enter elements of first set: "))
s2 = set(input("Enter elements of second set: "))

if s1.issubset(s2):
    print("First set is a subset of second set")
else:
    print("First set is NOT a subset of second set")
```

<h3>3) Write a program to remove duplicates from a list using sets.

```python
lst = [1, 2, 3, 2, 4, 1, 5]

unique_list = list(set(lst))

print("List without duplicates:", unique_list)
```

<h3>4) Write a program to find elements that are present in the first set but not in the second set.

```python
s1 = set(input("Enter elements of first set: "))
s2 = set(input("Enter elements of second set: "))

result = s1 - s2

print("Elements in first set but not in second:", result)
```

<h3>5) Write a program to check if two sets are disjoint (have no common elements).

```python
s1 = set(input("Enter elements of first set: "))
s2 = set(input("Enter elements of second set: "))

if s1.isdisjoint(s2):
    print("Sets are disjoint")
else:
    print("Sets are NOT disjoint")
```

<h2>Question Set B

<h2>1) Write a program to find symmetric difference between two sets.

```python
s1 = set(input("Enter elements of first set: "))
s2 = set(input("Enter elements of second set: "))

sym_diff = s1 ^ s2

print("Symmetric Difference:", sym_diff)
```

<h3>2) Write a program to add multiple elements to a set at once from user input.

```python
s = set()

elements = input("Enter elements to add (no spaces): ")

for e in elements:
    s.add(e)

print("Updated set:", s)
```

<h3>3) Write a program to find the length of a set and check if a specific element exists in it.

```python
s = set(input("Enter elements of the set: "))

print("Length of set:", len(s))

element = input("Enter element to check: ")

if element in s:
    print("Element exists in the set")
else:
    print("Element does not exist in the set")
```

<h2>Question Set C

<h3>1) Write a program that takes two sentences as input and finds:
● Common words in both sentences
● Words unique to the first sentence
● Words unique to the second sentence

```python
s1 = input("Enter first sentence: ")
s2 = input("Enter second sentence: ")

set1 = set(s1.split())
set2 = set(s2.split())

print("Common words:", set1 & set2)
print("Words unique to first sentence:", set1 - set2)
print("Words unique to second sentence:", set2 - set1)
```

<h3>2) Write a program to perform all set operations (union, intersection, difference, symmetric
difference) on three sets.

```python
s1 = set(input("Enter elements of first set: "))
s2 = set(input("Enter elements of second set: "))
s3 = set(input("Enter elements of third set: "))

print("Union:", s1 | s2 | s3)
print("Intersection:", s1 & s2 & s3)
print("Difference (s1 - s2 - s3):", s1 - s2 - s3)
print("Symmetric Difference:", s1 ^ s2 ^ s3)
```
