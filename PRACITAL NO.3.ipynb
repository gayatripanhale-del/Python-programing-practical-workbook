<h2>Question Set A

<h3>1) Write a program to find the maximum and minimum element in a tuple without using
max() and min() functions.

```python
t = (10, 5, 20, 3, 15)

max_val = t[0]
min_val = t[0]

for num in t:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("Maximum element:", max_val)
print("Minimum element:", min_val)
```

<h3>2) Write a program to count the occurrence of each element in a tuple.
Example: For tuple (1, 2, 3, 2, 1, 3, 2), output should show 1:2, 2:3, 3:2

```python
t = (1, 2, 3, 2, 1, 3, 2)

for i in t:
    count = 0
    for j in t:
        if i == j:
            count = count + 1
    print(i, ":", count)
```

<h3>3) Write a program to convert a tuple of strings into a single string.
Example: ("Hello", "World", "Python") should become "HelloWorldPython"

```python
t = ("Hello", "World", "Python")

result = ""

for word in t:
    result = result + word

print("Single string:", result)
```

<h3>4) Write a program to find the sum of all even numbers in a tuple.

```python
t = (1, 2, 3, 4, 5, 6, 8)

sum_even = 0

for num in t:
    if num % 2 == 0:
        sum_even = sum_even + num

print("Sum of even numbers:", sum_even)
```

<h3>5) Write a program to remove duplicates from a tuple (return a new tuple with unique
elements).

```python
t = (1, 2, 3, 2, 1, 4, 3)

unique = []

for num in t:
    if num not in unique:
        unique.append(num)

unique_tuple = tuple(unique)

print("Tuple without duplicates:", unique_tuple)
```

<h2>Question Set B

<h3>1) Write a program to find the second largest element in a tuple.

```python
t = (10, 5, 20, 8, 15)

largest = t[0]
second_largest = t[0]

for num in t:
    if num > largest:
        second_largest = largest
        largest = num
    elif num != largest and num > second_largest:
        second_largest = num

print("Second largest element:", second_largest)
```

<h3>2) Write a program to sort a tuple of tuples based on the second element.
Example: ((1, 5), (3, 2), (2, 8)) should become ((3, 2), (1, 5), (2, 8))

```python
t = ((1, 5), (3, 2), (2, 8))

sorted_list = list(t)

for i in range(len(sorted_list)):
    for j in range(i + 1, len(sorted_list)):
        if sorted_list[i][1] > sorted_list[j][1]:
            sorted_list[i], sorted_list[j] = sorted_list[j], sorted_list[i]

sorted_tuple = tuple(sorted_list)

print("Sorted tuple:", sorted_tuple)
```

<h3>3) Write a program to check if all elements in a tuple are unique.</h3>

```python
t = (1, 2, 3, 4, 5)

unique = True

for i in range(len(t)):
    for j in range(i + 1, len(t)):
        if t[i] == t[j]:
            unique = False

if unique:
    print("All elements are unique")
else:
    print("Elements are not unique")
```

<h3>4) Write a program to swap the first and last element of a tuple.

```python
t = (10, 20, 30, 40, 50)

lst = list(t)

lst[0], lst[-1] = lst[-1], lst[0]

new_tuple = tuple(lst)

print("Tuple after swapping:", new_tuple)
```

<h3>1) Write a program to find common elements between two tuples.</h3>

```python
t1 = (1, 2, 3, 4, 5)
t2 = (4, 5, 6, 7)

common = []

for i in t1:
    for j in t2:
        if i == j and i not in common:
            common.append(i)

common_tuple = tuple(common)

print("Common elements:", common_tuple)
```

<h3>2) Write a program to create a tuple of squares of numbers from 1 to n.
Example: If n=5, output should be (1, 4, 9, 16, 25)

```python
n = int(input("Enter value of n: "))

squares = []

for i in range(1, n + 1):
    squares.append(i * i)

square_tuple = tuple(squares)

print("Tuple of squares:", square_tuple)
```
