<h2>Question Set A

<h3>1) Write a program to create a dictionary with roll number as key and name as value for 5
students. Display all key-value pairs.

```python
students = {}

for i in range(5):
    roll = int(input("Enter roll number: "))
    name = input("Enter name: ")
    students[roll] = name

print("Student Details:")
for roll, name in students.items():
    print(roll, ":", name)
```

<h3>2) Write a program to count the frequency of each word in a given sentence using a
dictionary.Example: "hello world hello" should give {'hello': 2, 'world': 1}

```python
sentence = input("Enter a sentence: ")

words = sentence.split()
freq = {}

for word in words:
    if word in freq:
        freq[word] = freq[word] + 1
    else:
        freq[word] = 1

print("Word frequency:", freq)
```

<h3>3) Write a program to merge two dictionaries entered by the user.

```python
##### dict1 = {}
dict2 = {}

n1 = int(input("Enter number of elements in first dictionary: "))
for i in range(n1):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict1[key] = value

n2 = int(input("Enter number of elements in second dictionary: "))
for i in range(n2):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict2[key] = value

merged = dict1.copy()
merged.update(dict2)

print("Merged Dictionary:", merged)
```

<h3>4) Write a program to find the key with the maximum value in a dictionary.

```python
data = {'a': 10, 'b': 25, 'c': 15}

max_key = list(data.keys())[0]
max_value = data[max_key]

for key in data:
    if data[key] > max_value:
        max_value = data[key]
        max_key = key

print("Key with maximum value:", max_key)
```

<h2>Question Set B

<h3>1) Write a program to remove a key from a dictionary and handle the case when key doesn't
exist.

```python
data = {'a': 10, 'b': 20, 'c': 30}

key = input("Enter key to remove: ")

if key in data:
    del data[key]
    print("Key removed successfully")
else:
    print("Key does not exist")

print("Updated dictionary:", data)
```

<h3>2) Write a program to sort a dictionary by its values in descending order.

```python
data = {'a': 10, 'b': 50, 'c': 30}

sorted_items = sorted(data.items(), key=lambda x: x[1], reverse=True)

sorted_dict = {}
for key, value in sorted_items:
    sorted_dict[key] = value

print("Dictionary sorted by values (descending):", sorted_dict)
```

<h3>3) Write a program to create a dictionary from two lists - one containing keys and another
containing values.

```python
keys = ['a', 'b', 'c']
values = [10, 20, 30]

result = {}

for i in range(len(keys)):
    result[keys[i]] = values[i]

print("Created dictionary:", result)
```

<h2>Question Set C

<h3>1) Write a program to find all keys with the same value in a dictionary.</h3>

```python
data = {'a': 10, 'b': 20, 'c': 10, 'd': 30, 'e': 20}

value = int(input("Enter value to find keys: "))

keys = []

for k in data:
    if data[k] == value:
        keys.append(k)

if keys:
    print("Keys with value", value, ":", keys)
else:
    print("No keys found with value", value)
```

<h3>2) Write a program to create a dictionary where keys are numbers from 1 to n and values are
their cubes.

```python
n = int(input("Enter value of n: "))

cubes = {}

for i in range(1, n + 1):
    cubes[i] = i ** 3

print("Dictionary of cubes:", cubes)
```
