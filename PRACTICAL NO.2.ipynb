<h3>2) Write a program to check if a string is a palindrome (reads the same forwards and
backwards).

```python
s = input("Enter a string: ")

rev = ""
for ch in s:
    rev = ch + rev

if s == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

<h3>3) Write a program to reverse a string without using built-in reverse functions.

```python
s = input("Enter a string: ")

rev = ""
for ch in s:
    rev = ch + rev

print("Reversed string:", rev)
```

<h3>4) Write a program to count the frequency of each character in a string.
Example: For "hello", output should be h:1, e:1, l:2, o:1

```python
s = input("Enter a string: ")

for ch in s:
    if ch != " ":
        count = 0
        for c in s:
            if ch == c:
                count = count + 1
        print(ch, ":", count)
```

<h3>1) Write a program to find the first non-repeating character in a string.

```python
s = input("Enter a string: ")

found = False
for ch in s:
    count = 0
    for c in s:
        if ch == c:
            count = count + 1
    if count == 1:
        print("First non-repeating character:", ch)
        found = True
        break

if not found:
    print("No non-repeating character")
```

<h3>2) Write a program to check if two strings are anagrams (contain same characters in different
order).

```python
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

if len(s1) != len(s2):
    print("Not Anagrams")
else:
    is_anagram = True
    for ch in s1:
        if s1.count(ch) != s2.count(ch):
            is_anagram = False
            break

    if is_anagram:
        print("Anagrams")
    else:
        print("Not Anagrams")
```

<h3>3) Write a program to convert a string to title case without using the title() method.
Example: "python programming" should become "Python Programming"

```python
s = input("Enter a string: ")

result = ""
new_word = True

for ch in s:
    if ch == " ":
        result = result + ch
        new_word = True
    else:
        if new_word:
            result = result + ch.upper()
            new_word = False
        else:
            result = result + ch.lower()

print("Title Case:", result)
```

<h3>4) Write a program to replace all occurrences of a substring with another substring in a given
string.

```python
s = input("Enter main string: ")
old = input("Enter substring to replace: ")
new = input("Enter new substring: ")

result = ""
i = 0

while i < len(s):
    if s[i:i+len(old)] == old:
        result = result + new
        i = i + len(old)
    else:
        result = result + s[i]
        i = i + 1

print("Updated string:", result)
```

<h3>1) Write a program to find the longest word in a sentence.

```python
sentence = input("Enter a sentence: ")

words = sentence.split()

longest = words[0]

for word in words:
    if len(word) > len(longest):
        longest = word

print("Longest word:", longest)
print("Length:", len(longest))
```

<h3>2) Write a program to check if a string contains only digits, only alphabets, or both.

```python
s = input("Enter a string: ")

has_alpha = False
has_digit = False

for ch in s:
    if ch.isalpha():
        has_alpha = True
    elif ch.isdigit():
        has_digit = True

if has_alpha and has_digit:
    print("String contains both alphabets and digits")
elif has_alpha:
    print("String contains only alphabets")
elif has_digit:
    print("String contains only digits")
else:
    print("String contains other characters")
```
