<h3>1) Write a function to find the maximum of three numbers.

```python
def max_of_three(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c


print(max_of_three(10, 25, 15))
```

<h3>2) Write a function to check if a number is prime or not.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True


print(is_prime(7))
print(is_prime(10))
```

<h3>3) Write a function to calculate the factorial of a number (both iterative and recursive).

```python
def factorial_iterative(n):
    fact = 1
    for i in range(1, n + 1):
        fact = fact * i
    return fact


print(factorial_iterative(5))
```

```python
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)


print(factorial_recursive(5))
```

<h3>4) Write a function that accepts a list and returns a new list with unique elements.]

```python
def unique_list(lst):
    result = []
    for item in lst:
        if item not in result:
            result.append(item)
    return result


print(unique_list([1, 2, 3, 2, 1, 4]))
```

<h3>5) Write a function to calculate the sum of all even numbers in a list.</h3> 

```python
def sum_even(lst):
    total = 0
    for num in lst:
        if num % 2 == 0:
            total = total + num
    return total


print(sum_even([1, 2, 3, 4, 5, 6]))
```

<h3>1) Write a function to convert temperature from Celsius to Fahrenheit and vice versa.

```python
def celsius_to_fahrenheit(c):
    return (c * 9 / 5) + 32

def fahrenheit_to_celsius(f):
    return (f - 32) * 5 / 9


print("Celsius to Fahrenheit:", celsius_to_fahrenheit(25))
print("Fahrenheit to Celsius:", fahrenheit_to_celsius(77))
```

<h3>2) Write a function that accepts a string and returns the count of vowels and consonants.

```python
def count_vowels_consonants(s):
    vowels = 0
    consonants = 0

    for ch in s:
        if ch.isalpha():
            if ch in "aeiouAEIOU":
                vowels = vowels + 1
            else:
                consonants = consonants + 1

    return vowels, consonants


v, c = count_vowels_consonants("python")
print("Vowels:", v)
print("Consonants:", c)
```

<h3>3) Write a function that accepts variable number of arguments and returns their average.

```python
def average(*numbers):
    total = 0
    count = 0

    for num in numbers:
        total = total + num
        count = count + 1

    return total / count


print("Average:", average(10, 20, 30, 40))
```

<h3>4) Write a lambda function to find the maximum of three numbers.

```python
max_of_three = lambda a, b, c: a if (a >= b and a >= c) else (b if b >= a and b >= c else c)

print(max_of_three(10, 25, 15))
```

<h3>1) Write a recursive function to calculate the sum of digits of a number.

```python
def sum_of_digits(n):
    if n == 0:
        return 0
    return (n % 10) + sum_of_digits(n // 10)


print("Sum of digits:", sum_of_digits(130))
```

<h3>2) Write a function that accepts a list of numbers and returns a dictionary with count of even
and odd numbers.
Example: [1,2,3,4,5] should return {'even': 2, 'odd': 3}

```python
def count_even_odd(lst):
    result = {'even': 0, 'odd': 0}

    for num in lst:
        if num % 2 == 0:
            result['even'] = result['even'] + 1
        else:
            result['odd'] = result['odd'] + 1

    return result


print(count_even_odd([1, 2, 3, 4, 5]))
```
