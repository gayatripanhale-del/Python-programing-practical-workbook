<h3>1) Write a Python Program to Calculate the Average of Numbers in a Given List.

<h3>2) Write a program which accepts 6 integer values and prints "DUPLICATES" if any of the
values entered are duplicates otherwise it prints "ALL UNIQUE".Example: Let 6 integers are (32, 10, 45, 90, 45, 6) then output "DUPLICATES" to be printed.</h3>

```python
numbers = []
for i in range(6):
    num = int(input(f"Enter integer {i+1}: "))
    numbers.append(num)

if len(numbers) != len(set(numbers)):
    print("DUPLICATES")
else:
    print("ALL UNIQUE")
```

<h3>4) Write a program which finds sum of digits of a number.
Example: n=130 then output is 4 (1+3+0).</h3>

```python
n = int(input("Enter a number: "))

sum = 0

while n > 0:
    digit = n % 10
    sum = sum + digit
    n = n // 10

print("Sum of digits =", sum)
```

<h3>5) Write a program which prints Fibonacci series of a number.

```python
n = int(input("Enter number of terms: "))

a = 0
b = 1

for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```

<h3>1) Write a program which accept an integer value 'n' and display all prime numbers till 'n'.

```python
n = int(input("Enter a number: "))

print("Prime numbers till", n, "are:")

for num in range(2, n + 1):
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
```

<h3>4) Write a program to reverse a given number.</h3>

```python
n = int(input("Enter a number: "))

reverse = 0

while n > 0:
    digit = n % 10
    reverse = reverse * 10 + digit
    n = n // 10

print("Reversed number:", reverse)
```

<h3>1) Write a Sequential search function which searches an item in a sorted list. The function
should return the index of element to be searched in the list.</h3>

```python
def search(lst, item):
    i = 0
    while i < len(lst):
        if lst[i] == item:
            return i
        i = i + 1
    return -1


numbers = [10, 20, 30, 40, 50]

item = int(input("Enter element to search: "))

index = search(numbers, item)

if index != -1:
    print("Element found at index:", index)
else:
    print("Element not found")
```
